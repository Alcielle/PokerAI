```python
import math
import random
import numpy as np

x_train = np.zeros((60000,3))
y_train = np.zeros(60000)

x_test = np.zeros((10000,3))
y_test = np.zeros(10000)

action = 0

for i in range(70000):
    j = random.randint(1,52)
    handscore = random.randint(0, 500)
    random.seed(j)
    opptscore = random.randint(0,500)
    random.seed(j+1)
    
    w2 = random.random()

    opptscore = opptscore * w2

    gamescore = handscore + opptscore

    if gamescore<200:
      action = 0 #fold/check
    elif gamescore<500:
      action = 1 #call/check
    else:
      action = 2 #bet
    if i<10000:
      x_test[i][0] = handscore
      x_test[i][1] = opptscore
      x_test[i][2] = action
      y_test[i] = action
    else:
      x_train[i-10000][0] = handscore
      x_train[i-10000][1] = opptscore
      x_train[i-10000][2] = action
      y_train[i-10000] = action    
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Activation
from keras.layers import Dropout
```

```python
model = Sequential()
model.add(LSTM(3,input_shape=(3,1),return_sequences=False))
model.add(Dropout(0.5))
model.add(Dense(100,kernel_initializer='normal', activation='relu'))
model.add(Dense(3,kernel_initializer='normal',activation='softmax'))
model.compile(loss='sparse_categorical_crossentropy',optimizer ='adam',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=20,batch_size=60,validation_split=0.05,verbose=1)
```

```python
scores = model.evaluate(x_test, y_test, verbose=1, batch_size=5)
```

```python
import matplotlib.pyplot as plt

predict = model.predict(x_test)
print(y_test)
```
